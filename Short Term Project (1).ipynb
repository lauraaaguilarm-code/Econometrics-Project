```r
library(tidyverse)
library(haven)
library(dplyr)
install.packages('wooldridge')
```

```r
library("wooldridge")
data(affairs, package = "wooldridge")
head(affairs)
```

```r
affairs_reg1 <- lm(affair ~ educ, data = affairs)
summary(affairs_reg1)
```

```r
affairs_reg2 <- lm(affair ~ educ + relig, data = affairs)
summary(affairs_reg2)
```

```r
library(ggplot2)

ggplot(affairs, aes(x = factor(affair), y = educ)) +
  geom_boxplot(fill = "red") +
  labs(title = "Years of Education by Affair Status",
       x = "Had an Affair (0 = No, 1 = Yes)",
       y = "Years of Education") 
```
